```python
# 강아지 사진
# "dog.jpg"
# YOLOv8
from ultralytics import YOLO
import cv2

model = YOLO("yolov8s.pt")  # YOLOv8의 s모델 쓰자 (참고로 없으면 다운로드 / 있으면 씀)

frame = cv2.imread("dog.jpg")

detect = model(frame)
# YOLOv8은 plot() 함수를 통해 box, class_name, confidence까지 잘 나오게 함수화
cv2.imshow("Here", detect[0].plot())
# cv2.imshow("frame", frame)
cv2.waitKey(0)

cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# 실시간 웹캠을 통해 YOLOv8 모델 돌려보기
import cv2
from ultralytics import YOLO

model = YOLO("yolov8s.pt")
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    exit()
try:
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        detect = model(frame)
        cv2.imshow("frame", detect[0].plot())

        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
finally:
    cap.release()
    cv2.destroyAllWindows()
    cv2.waitKey(1)
```
