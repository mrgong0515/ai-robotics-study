```python
import cv2          # 이미지 불러오기(실시간 비디오 등)
import torch        # yolov5 모델 불러오기
import numpy as np  # 결과들을 다룰 때(array)
import yaml         # data의 class이름 등 정보 로드

cap = cv2.VideoCapture(0)
model = torch.hub.load("ultralytics/yolov5", "yolov5s")

with open("yolov5/data/coco.yaml", "r", encoding="UTF-8") as f:
    data = yaml.full_load(f)

if not cap.isOpened():
    print("웹 캡을 연결해주세요")
else:
    print("웹 캠 연결 완료")
    while True:
        ret, frame = cap.read()
        if not ret:
            print("프레임을 읽을 수 없습니다")
            break

        result = model(frame)
        detect = result.xyxyn[0].numpy()

        x_shape, y_shape = frame.shape[1], frame.shape[0]
        # detect는 현재 감지된 객체만큼의 길이를 갖고 있음
        for d in detect:
            x1 = int(d[0] * x_shape)
            y1 = int(d[1] * y_shape)
            x2 = int(d[2] * x_shape)
            y2 = int(d[3] * y_shape)
            conf = d[4]
            name = data["names"][int(d[5])]
            text = f"{name}:{conf:.2f}"
        
            color = (0, 0, 255)
            thickness = 4
            cv2.putText(frame, text, (x1, y1 + 30), cv2.FONT_HERSHEY_PLAIN, 2, color, 2)
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, thickness)
        
        cv2.imshow("frame", frame)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            # frame하나 출력 후 1ms 대기
            # 혹시나 q 버튼 눌려지면 통과 
            break
cap.release() # 카메라 점유 해제
cv2.destroyAllWindows()
cv2.waitKey(1)
```
