```python
# 자료구조
# 데이터를 효율적으로 다루기 위한 방법
# 1) 선형
# - 배열, 연결 리스트, 스택, 큐
# 2) 비선형
# - 트리, 그래프
```

```python
# 사실...Python에서 List를 쓰면
# 배열을 구현 안해도 된다

class Array:
    def __init__(self):
        self.data = []

    def append(self, data):
        self.data.append(data)

    def get(self, index):
        return self.data[index]

    def update(self, index, data):
        self.data[index] = data  # 수정 로직 구현!!
                                 # 리스트 값 추가 : append, insert
                                 # 수정 -> []
                                 # 삭제 -> del, remove
        
    def pop(self):
        self.data.pop() # 가장 마지막 요소 삭제
        
    def remove(self, index):
        del self.data[index] # 특정 인덱스 삭제
    def __str__(self):
        return str(self.data) #모든 요소 출력 [10, 20, 30]
```

```python
arr = Array()
arr.append(10)
arr.append(20)
arr.append(30)
print(arr.get(1)) # 1번 인덱스의 값을 반환
print(arr)
arr.update(1,40)
print(arr)
arr.pop
print(arr)
arr.remove(0)
print(arr)
```

```python
# Queue(큐)
# First In First Out
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.head = None
        self.tail = None
        self.cnt = 0

    def push(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = self.tail = new_node
        else:
            current = new_node
            self.tail.next = current
            self.tail = current

        self.cnt += 1

    def empty(self):
        return self.head is None

    def pop(self):
        if self.empty():
            self.tail = None
            self.cnt = 0
            return -1
            
        current = self.head
        self.head = current.next
        self.cnt -= 1
        return current.data

    def size(self):
        return self.cnt

    def front(self):
        return -1 if self.empty() else self.head.data
    def back(self):
        return -1 if self.empty() else self.tail.data
```

```python
q = Queue()

print("empty:", q.empty())

q.push(10)
q.push(20)
q.push(30)

print("size:", q.size())
print("front:", q.front())
print("back:", q.back())

print("pop:", q.pop())
print("size:", q.size())
print("front:", q.front())
print("back:", q.back())

print("pop:", q.pop())
print("pop:", q.pop())
print("pop:", q.pop())  # 비어있으면 -1
```

```python
# Dictionary
# 자료 구조 중 하나

# python -> C

# CSV(comma separated value)
# 이름,나이,MBTI
# 나예호,20,ENFJ
# 수지,30,INFP
# -> 공백 포함 42byte
# 장점 : 대량 데이터를 다루기 용이
# 단점 : 데이터의 정보 확인 어려움

# XML(eXtensible Markup Language)
# <이름>"나예호"</이름><나이>20</나이><MBTI>"ENFJ"</MBTI>
# <이름>"수지"</이름><나이>30</나이><MBTI>"INFP"</MBTI>
# -> 공백 포함 109byte
# 장점 : 메타데이터 포함 -> 정보를 확인하기 쉽다
# 단점 : 비용이 많이 든다

# JSON(JavaScript Object Notation)
# {"이름":"나예호", "나이":20, "MBTI":"ENFJ"},
# {"이름":"수지", "나이":30, "MBTI":"INFP"}
# -> 공백 포함 86byte
# 장점 : CSV 대비 정보 확인 쉽다, XML대비 비용 적게 든다
# 단점 : CSV 대비 비용 많이 든다, XML대비 정보 확인 어렵다
```

```python
# Python에서 Jason을 Dictionary라고 부른다

dic_sw = {"이름":"공성우", "나이":30, "MBTI":"ENTJ"}
dic_sy = {"이름":"권수연", "나이":26, "MBTI":"INTJ"}
```

```python
print(dic_sw)
print(dic_sy)
# dictionary 값 확인
# :을 기준으로 key-value 쌍으로 구성
# 딕셔너리명[키값]
print(dic_sw["나이"])
# print(dic_sw["나"]) 없는 key 입력 시 Error
print(dic_sw.get("나이"))
print(dic_sw.get("나"))

dic_sw["혈액형"] = "O"
print(dic_sw)
dic_sw["혈액형"] = "O"
print(dic_sy)

# 딕셔너리 값 수정
# 딕셔너리[키값] = 밸류값
dic_sw["나이"] -= 1
print(dic_sw)
```

```python
dic_list = [dic_sw, dic_sy]
# 나이의 총합
age_total = 0

for dic in dic_list:
    age_total += dic["나이"]

print(age_total)
print(dic_list)
```
