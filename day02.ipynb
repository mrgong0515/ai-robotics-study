```python
# 정렬 알고리즘
# 정렬은 왜 할까요?
# Bubble Sort
# 두 인접한 요소를 비교하여
# 요소를 치환
array = [7, 45, 12, 82, 25]
print(f"정렬 전 : {array}")

for j in range(4, 0, -1):
    for i in range(j):
        if array[i] > array[i + 1]:
            temp = array[i]
            array[i] = array[i + 1]
            array[i + 1] = temp
    
print(f"정렬 후 : {array}") # [7, 12, 25, 45, 82]
```

```python
# 선택 정렬(Selection Sort)

arr = [98, 7, 70, 13, 24]

# 오름차순 정렬
# 가장 큰 값이 존재하는 인덱스의 정보를 탐색하여
# 하나의 step이 종료되면
# 치환!!

for j in range(5, 1, -1):
    max_index = 0 # 가장 큰 요소의 위치 정보

    for i in range(j):
        if arr[i] > arr[max_index]:
            max_index = i

    temp = arr[max_index]
    arr[max_index] = arr[j - 1]
    arr[j - 1] = temp

print(arr)
```

```python
# 탐색(Search)

# 이진 탐색(Binary Search) - 1
list31 = [1, 7, 16, 25, 30, 33, 41, 66, 78, 90]
search = 78

low_index = 0
high_index = 9
mid_index = (low_index + high_index) // 2

while True:
    if list31[mid_index] > search:
        high_index = mid_index - 1
    elif list31[mid_index] < search:
        low_index = mid_index + 1
    else:
        print(f"찾고자 하는 값은 : index : {mid_index} 위치에 있습니다")
        break
    mid_index = (low_index + high_index) // 2
```

```python
# 탐색(Search)

# 이진 탐색(Binary Search) -3
list31 = [1, 7, 16, 25, 30, 33, 41, 66, 78, 90]
search = 88

low_index = 0
high_index = 9
mid_index = (low_index + high_index) // 2
# 찾고자 하는 값이 없을 때
# 해당 값은 존재하지 않습니다
is_find = False

while low_index <= high_index: # 탐색 값이 없는 상황을 예외
    if list31[mid_index] > search:
        high_index = mid_index - 1
    elif list31[mid_index] < search:
        low_index = mid_index + 1
    else:
        print(f"찾고자 하는 값은 : index : {mid_index} 위치에 있습니다")
        is_find = True
        break
    mid_index = (low_index + high_index) // 2
    
if not is_find:
    print("해당 값은 존재하지 않습니다.")
```

```python
# 완전수 (Perfect Number)
# 오늘의 함수 예제 목표 : 완전수 구하기 !!
# 자기 자신을 제외한 약수의 합이
# 자기 자신과 같은 수

# ex. 6 -> 1 2 3 6 : 첫 번째 완전수
#     8 -> 1 2 4 8 : 완전수 X
```

```python
# 1) is_divisor(num1, num2) -> bool
# 만약, num2가 num1의 약수라면 True
# 약수가 아니라면 False 반환하는 함수 정의

def is_divisor(num1, num2):
    # Logic
    if num1 % num2 == 0:
        return True
    else:
        return False
        
print(is_divisor(10, 2)) #True
print(is_divisor(10, 3)) # False
print(is_divisor(40, 8))
```

```python
# 2) print_divisors(num)
# print_divisors(10)
# 10의 약수 : 1 2 5 10
def print_divisors(num):
    print(f"{num}의 약수 : ", end = "")
    for i in range(1, num // 2 + 1):
        if is_divisor(num, i): # 내가 정의한 다른 함수를 호출해서 사용 가능!!
            print(i, end = " ")
    print(num)
    
print_divisors(10) #10의 약수 : 1 2 5 10
print_divisors(28) # 28의 약수 : 1 2 4 7 14 28
```

```python
# 3) get_sum_of_divisors(num)
# 약수의 총합을 구하는 함수
def get_sum_of_divisors(num):
    total = 0

    for i in range(1, num // 2 + 1):
        if is_divisor(num, i):
            total += i
    total += num
    return total
print(get_sum_of_divisors(10)) # 18
print(get_sum_of_divisors(28)) # 56
```

```python
# 4) is_perfect_number(num)
# num이 완전수라면 True, 아니라면 False 반환 !!

# hint 1) 처음부터 구하기보단 지금까지 정의된 함수 활용하기
# hint 2) 6은 완전수다, 28도 완전수다
# hint 3) 완전수인 수들의 약수의 총합은 ??

def is_perfect_number(num):
    if get_sum_of_divisors(num) == num * 2: #약수의 총합이 자기 자신의 두배라면 완전수다 라고 정의
        return True
    else:
        return False
        
print(is_perfect_number(6)) # True
print(is_perfect_number(10)) # False
print(is_perfect_number(28)) # True
```

```python
# 5)
# 첫 번째 완전수 : 6
# 두 번째 완전수 : 28
# 세 번째 완전수 : 100 ~ 1000
# 네 번째 완전수 : 1000 ~ 10_000

for i in range(1, 10):
    if is_perfect_number(i):
        print(i)

for i in range(10, 100):
    if is_perfect_number(i):
        print(i)

for i in range(100, 1000):
    if is_perfect_number(i):
        print(i)

for i in range(1000, 10000):
    if is_perfect_number(i):
        print(i)
```
