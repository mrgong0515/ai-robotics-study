```python
# 예외처리
# 실행 시 Error Class
try:
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))

    
except:
    print("정수를 입력해주세요")

# try ~ except (예외가 날 것 같은 실행문장을 감싼다. 나눌 수 없을 때 에러가 나지 않고, except로 빠짐)
try:
    if num1 % num2 == 0:
        print(f"{num2}는 {num1}의 약수입니다")
    else:
        print(f"{num2}는 {num1}의 약수가 아닙니다")
except:
    print("0으로 나눌 수 없습니다")

print("프로그램 종료")
```

```python

try:
    num1 = int(input("첫 번째 정수 입력 >>"))
    num2 = int(input("두 번째 정수 입력 >>"))

    if num1 % num2 == 0:
        print(f"{num2}는 {num1}의 약수입니다")
    else:
        print(f"{num2}는 {num1}의 약수가 아닙니다")

except ValueError: # 정수가 아닌 int형이 아닌 무언가 입력 시
    print("정수를 입력해주세요")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다")
except:
    print("에러 발생!!") # 내가 모르는 에러 발생시

finally: # 예외 발생 상관 없이 실행되는 부분
    print("프로그램 종료")
```

```python
# 근로소득 자료를 토대로 다음의 2개 항목을 출력하시오
# 1) 중앙값 (50% 구간의 평균 소득)
search = "50%"
with open("salary.csv", "r", encoding="euc-kr") as f:
    reader = csv.DictReader(f)
    for row in reader:
        if search in row["구분"] :
            print(row)
            earned_income = int(row["근로소득금액"]) / int(row["인원"])
            final_tax = int(row["결정세액"]) / int(row["인원"])
            print(f"{search} 구간 근로소득 : {earned_income * 10000 : .0f}만원")
            print(f"{search} 구간 결정세액 : {final_tax * 10000 : .0f}만원")

# 2) 평균값(구간별 인원수, 구간별 소득금액)
total_popular = 0
total_salary = 0
with open("salary.csv", "r", encoding="euc-kr") as f:
    reader = csv.DictReader(f)
    for row in reader:
        total_popular += int(row["인원"])
        total_salary += int(row["근로소득금액"])
print(f"평균 근로소득 : {total_salary / total_popular * 10000 : .0f}만원")
```

```python
df_poke = pd.read_csv("pokemon.csv")
# df_poke.info()
df_poke.describe() # 수치 데이터들 확인
df_poke.head()
# DataFrame 기본 인덱싱 및 슬라이싱
# 인덱싱
# 데이터프레임[컬럼명1] -> 1차원 시리즈
df_poke["Name"]
# 데이터프레임[[컬럼명1],[컬럼명2]] -> 2차원 시리즈
df_poke[["Name", "Total"]]
```

```python
# 조건에 맞는 데이터 추출
# loc (location)
# 데이터프레임.loc[행 인덱싱 값]
df_poke.loc[0] # 0번 인덱스의 data -> 1차원 시리즈

# 데이터프레임.loc[행 인덱싱 값, 열 인덱싱 값]
df_poke.loc[0, "Total"] # 0번 인덱스의 Total 컬럼 data
df_poke.loc[0, "Name"] # 0번 인덱스의 Name 컬럼 data

# 데이터프레임.loc[행 슬라이싱]
df_poke.loc[0 : 2] # 0번 인덱스 ~ 2번 인덱스

# 데이터프레임.loc[행 슬라이싱, 열 슬라이싱]
df_poke.loc[0 : 2,"Name":"Total"] # 0번 인덱스 ~ 2번 인덱스, Name 컬럼 ~ Total 컬럼
```

```python
# Boolean Indexing
# 포켓몬 Total이 400 이상인 데이터들만 확인!!

# 데이터프레임[행 인덱싱]
df_poke["Total"] >= 400
df_poke[df_poke["Total"] >= 400]

df_poke.loc[df_poke["Total"] >= 400]

df_poke[(df_poke["Total"] >= 400) & (df_poke["Generation"] == 1)] # 1세대 포켓몬과 토탈 400이상인 데이터들만 확인 # 두가지 모두 충족한 데이터만 추출
df_poke.loc[(df_poke["Total"] >= 400) & (df_poke["Generation"] == 1)] # 위와 같음 : 강사님 선호
```

```python
# Type1이 Fire인 포켓몬의 개수
df_poke.loc[df_poke["Type1"] == "Fire"]
# Type1이 Water인 포켓몬 중 Total 500이상인 포켓몬의 개수
df_poke.loc[(df_poke["Total"] >= 500) & (df_poke["Type1"] == "Water")]
# 개수는 출력시 나오는 rows를 확인하자!!

```

```python
# 특정 컬럼만 추출하여 복사

# 복사에 대해서 알아보자
# 얕은 복사
# 하나의 메모리 주소를 가리킴

list_snack = ["뿌뿌딸기맛", "뿌뿌메론맛", "뿌뿌초코맛"]
print(list_snack)

list_copy = list_snack
print(list_copy)
print(list_copy[0])

list_copy[0] = "뿌뿌바나나맛"
print(list_copy[0])

print(list_snack)
print(list_copy)
```

```python
# 리스트가 1차원이면 리스트명.copy()
# 좀 더 얕은 복사
# 원본은 그대로

list_copy2 = list_snack.copy()
print(list_copy2)
list_copy2[0] = "맛동산"
print(list_copy2)
print(list_snack)
```

```python
list_drink = ["ZICO", "솔의눈", [0, 1, 2]]
print(list_drink)

list_copy3 = list_drink.copy()
print(list_copy3)
list_copy3[0] = "데자와"
print(list_copy3)
print(list_drink)

list_copy3[2][0] = 100
print(list_copy3)
print(list_drink)  # Copy가 1차원이 아닐 때는 사용 위험성이 있다. & 문제 발생 가능성 있음.

# 얕은 복사를 하는 이유 -> 메모리 이점을 얻기 위해서
```

```python
# 깊은 복사!!
# 전체 Value를 다 복사!! 원본과 독립적으로

from copy import deepcopy
list_copy4 = deepcopy(list_drink)
print(list_copy4)

list_copy4[2][0] = 200
print(list_copy4)
print(list_drink)
```
