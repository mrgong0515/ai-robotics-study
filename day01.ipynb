```python
# 최소 동전의 문제
# 8700원
money = 8700
# 천원 : 8장, 백원 : 7개 -> 15개
# 오천원 : 1장, 천원 3장, 백원 7개 -> 11개
# 오천원 : 1장, 천원 3장, 오백원 1개, 백원 2개 -> 7개

# 코드를 작성할 때,
# 결과는 동일하되
# 가독성을 높히거나, 실행속도를 높히거나
# Refactoring

money_5000 = 1
money_1000 = 3
money_500 = 1
money_100 = 2
print(f"오천원 : {money_5000}장")
print(f"천원 : {money_1000}장")
print(f"오백원 : {money_500}개")
print(f"백원 : {money_100}개")
```

```python
# 3) 대입연산자
# 오른쪽의 '값'이 왼쪽 '변수'에 할당
salary = 4_000_000
# 임베디드 과정
# 월급 10% 상향!!
# 복합대입연산자
# =과 함께 산술연산자, 지수연산자
# +=, -=, =, /=, //=, %=, *= 
salary *= 1 + (0.1)
print(f"월급 : {salary}원")
```

```python
# 4) 비교연산자
# >, >=, <, <=, ==, !=
# 비교연산자의 결과는 항상 논리값이다.

print(3 > 5)
print(3 >= 5)
print(3 < 5)
print(3 > 5)
print(3 <= 5)
print(3 == 5)
print(3 != 5)
```

```python
# 5) 논리연산자
# not : 논리 결과를 부정 (T -> F, F -> T)

# 반에 불이 꺼져 있습니다.
# 만약, 반에 불이 꺼져 있으면 켜고
# 켜져 있으면 꺼주세요

isLight = False

isLight = not isLight

# and
# 논리값 and 논리값
# 둘 다 True 여야만 True가 나온다!!
# 1 * 1 = 1 (True and True -> True)
# 1 * 0 = 0 (True and False -> False)
# 0 * 1 = 0 (False and True -> True)
# 0 * 0 = 0 (False and False -> True)
# 만약, and 뒷 쪽에 반드시 실행되어야 하는 코드가 있을 경우
# 주의!! shortcut circuit
# and에서 앞 논리값이 False라면
# 뒷쪽은 Dead Code
print((3 > 5) and (10 > 5))
```

```python
# or
# 논리값 or 논리값
# 둘 다 False 일 때만, False가 나온다.
# 1 + 1 = 2 (True or True -> True)
# 1 + 0 = 1 (True or False -> True)
# 0 + 1 = 1 (False or True -> True)
# 0 + 0 = 0 (False or False -> False)
# 만약, or 뒷 쪽에 반드시 실행되어야 하는 코드가 있을 경우
# 주의!! shortcut circuit
# or에서 앞 논리값이 True라면
# 뒷쪽은 Dead Code
print((3 < 5) or (10 > 5))
```

```python
# 조건(삼항) 연산자
# 항 : 연산의 영향을 받는 것
# a + b

# 단항 연산자
# 이항 연산자
# 삼항 연산자

# a if 조건식 else b
# 조건식 : 논리값(True, False)
score = 90
grade = "A" if score >= 90 else "F"
print(grade)
```

```python
# 삼항 연산자를 활용하여 다음의 문제를 푸시오

# 키보드로 정수를 입력 받아 홀수인지 짝수인지 판별하시오

num = int(input("정수 입력 >>"))

result = "짝수" if num % 2 == 0 else "홀수"

print(f"{num}는(은) {result}입니다")
```

```python
# 키보드로 변수 num을 입력 받고 num이 3의 배수이면서 5의 배수이면 " 3과 5의 배수입니다"를 출력하고 아니라면 " 3과 5의 배수가 아닙니다"를 출력하시오
num = int(input("정수 입력 >>"))

if num % 3 == 0 and num % 5 == 0:
    print("3과 5의 배수입니다.")
else:
    print("3과 5의 배수가 아닙니다.")
```
