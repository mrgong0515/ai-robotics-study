```python
text = "The quick brown fox jumps over the lazy dog"
vowels = "aeiou"
count = 0

for char in text.lower():
    if char in vowels:
        count += 1

print(f"주어진 문자열에는 모음이 {count}개 있습니다.")
```

```python
# while문 사용 1부터 1000까지의 자연수 중 3의 배수의 합
result = 0
i = 1
while i <= 1000:
    if (i % 3 == 0):
        result += i
    i += 1
print(result)
```

```python
# A학급에 총 10명의 학생. for문 사용하여 A학급의 평균 점수를 구해보자
# [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
total = 0
for score in A:
    total += score

average = total / 10 # = len(A)

print(f"{average:.1f}")
```

```python
# 리스트 컴프리헨션 사용하기
# 리스트의 요소 중에서 홀수만 골라 2를 곱한 값을 result 리스트에 담는다.
numbers = [1, 2, 3, 4, 5]
result = []
for n in numbers:
    if n % 2 == 1:
        result.append(n * 2)

print(result)
```

```python
# 컴프리헨션 사용
numbers = [1, 2, 3, 4, 5]
result = [n * 2 for n in numbers if n % 2 == 1 ]
print(result)
```
