```python
# 캡슐화(Encapsulation)
# 사용자들이 접근하는 정보에 대한 제한

# 은행 클래스
class Bank:
    __money = 0 #전역 변수

    def deposit(self, money):
        self.__money += money
        
    def withdraw(self, money):
        self.__money -= money
        
    def show_me_the_money(self):
        print(f"잔액 : {self.__money}원")
```

```python
salary_account = Bank()

# 사용자가 접근하면 안되는 멤버에 대해
# 접근 제어(Access Modifier)
# 1) Public : money     -> 누구나 접근 가능!!
# 2) Protected : _money    -> 같은 패키지 혹은 상속 관계
# 3) Private : __money     -> 같은 클래스 내에서만 접근 가능!!
salary_account.money = 1_000_000_000

salary_account.deposit(1000)
salary_account.withdraw(500)

salary_account.show_me_the_money()
```

```python
# 포켓몬 VO 만들어 보자 -> 명세
# 클래스 이름 : Pokemon
# 멤버변수(field) : name, p_type, height, weight
# 멤버함수(function) : show_info()
# 이름 : 꼬부기, 타입 : 물, 키 : 0.5m, 몸무게 : 9.0kg
```

```python
class Pokemon:
    # 생성자 함수는 클래스 생성시 최초 호출되는 함수!!
    def __init__(self, name, p_type, height, weight):
        # init 안에 있는 멤버 변수는
        # 대체적으로 private(__) 속성으로 선언!!
        self.__name = name
        self.__p_type = p_type
        self.__height = height
        self.__weight = weight

    # 키만 변경할 수 있는 기능(함수)을 만들어주자
    # Setter 함수
    def set_height(self, height):
        self.__height = height
        
    # 몸무게를 반환하는 함수
    # Getter 함수
    def get_weight(self):
        return self.__weight
        
    # str함수의 특징 : 인스턴스의 이름을 부르면 내장되어있는 메모리가 불러짐
    # show info -> __str__
    # print 지우고 return 삽입
    def __str__(self):
        return f"이름 : {self.__name}, 타입 : {self.__p_type}, 키 : {self.__height}m, 몸무게 : {self.__weight}kg"
```

```python
# Music 클래스 설계
# field : title, artist
# __init__() 활용해서 설계!!
# 모든 field는 private
# 모든 field에 대해 Getter, Setter 정의
# __str__() 함수는 다음을 반환한다
# 제목 : Good Goodbye, 가수 : 화사
```

```python
# 상속(Inheritance)

# 호랑이, 코끼리, 아르마딜로, 랫서판다
# 위 4개의 공통적인 특징 추출 -> 포유류

# 포유류 클래스를 만들어보자
class Mammal:
    def __init__(self, weapon, height, weight):
        self.__weapon = weapon
        self.__height = height
        self.__weight = weight

    def attack(self):
        print(f"{self.__weapon}로 공격!!")

    def feed(self):
        print("젖을 먹인다")
```

```python
class Whale(Mammal):
    # 고래는
    # weapon, height, weight 이외에
    # feature라는 멤버변수 하나 추가하고 싶다
    def __init__(self, feature, weapon, height, weight):
        super().__init__(weapon, height, weight)
        self.__feature = feature

    def swim(self):
        print("헤엄~ 헤엄~")

    # 다형성(Polymorphism)
    # 부모 클래스의 함수를 자식 클래스에서 재정의
    # -> Overriding
    def attack(self):
        print("몸통 박치기")
```

```python
# 객체 지향 프로그래밍(OOP)
# 독립적인 객체 (Data, Logic)을 만들어
# 메세지를 통해 프로그램을 설계하는 패러다임

# 객체지향 프로그래밍 특징 4가지
# 추상화(Abstraction)
# 객체의 공통적인 특징을 추출하여 클래스를 설계
# 코끼리 -> 상아, 코가 길어요, 덩치가 크다
# 코끼리, 호랑이, 아르마딜로, 랫서 판다의 공통적인 특징인 포유류 자체를 추출하는 행위

# 포유류, 조류, 어류 -> 생물

# 캡슐화(Encapsulation)
# 사용자가 굳이 몰라도 되는 정보 (멤버변수, 멤버함수)를
# 접근 제어자(Access Modifier)를 통해 제어하는 행위
# 개발자가 의도한 기능만 공개하는 행위
#   ㄴ 정보 은닉 : 알면 안되는 정보(보안 이슈)를 감추는 행위

# 상속(Inheritance)
# 이미 만들어진 검증된 클래스를 물려 받아 새로운 클래스를 정의
# 물려 주는 클래스 : 부모 클래스, Super Class, 상위 클래스
# 물려 받는 클래스 : 자식 클래스, Sub Class, 하위 클래스
# 자식 클래스 공통 형질 추상화 -> 부모 클래스 정의

# 다형성(Polymorphism)
# 다양한 형태를 나타낼 수 있는 능력
# 함수 하나 호출 -> 여러 결과가 나온다
# 상속 받은 자식 클래스에서 Overriding (함수 재정의)

# 클래스의 다른 용도
# VO(Value Object) : 사용자 정의 객체
# 사용자가 원하는 자료형을 만든다 (마치 C언어의 구조체 처럼)
# __init__() -> 생성자 함수
# __str__() -> 인스턴스명 호출 시 반환되는 함수
# Getter, Setter 함수
```

```python
# 자료구조
# 1) 선형 자료구조
#  1. 연결 리스트 (Linked List)
# 2) 비선형 자료구조
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
# 1) 연결 리스트(Linked List)
class LinkedList():
    def __init__(self):
        self.head = None 

    def append(self, data):
        new_node = Node(data)

        # 만약, 아무런 노드가 없다면
        # 새로 들어온 노드를 head로 임명!!

        if self.head is None:
            self.head = new_node
        else: 
            
            # 맨 끝 노드까지 이동
            # 시작하는 부분 head
            # 주의) 연결 리스트에서 head는 항상 첫 노드를 가리켜야 한다
            # -> self.head에 대입 금지!!
            current = self.head

            while current.next is not None:
                current = current.next 

            current.next = new_node

    def display(self):
        # 10 -> 20 -> 30
        if self.head is None:
            print("연결 리스트가 비어 있습니다.")
            return

        current = self.head

        while current is not None:
            print(f"{current.data} -> ", end="") 
            current = current.next
        
        print("None")

    # 탐색
    def search(self, target):
        # target이 전 노드를 탐색하면서 있다면!!
        # index를 반환
        # 찾았다!! 40은 3번 index에 있습니다!!
        # 없다면 -> 찾을 수 없습니다

        current = self.head
        index = 0
        
        while current is not None:
            if current.data == target:
                print(f"찾았다!! {target}은 {index}번 index에 있습니다!!")
                break
            else:
                current = current.next
                index += 1

        print("찾을 수 없습니다.")

    # 수정
    def update(self, target, new_data):
        current = self.head
        while current:
            if current.data == target:
                current.data = new_data
                return

            current = current.next
     # 삭제 
    def delete(self, target):
        current = self.head

        if self.head.data == target:
            self.head = current.next
            return

        while current.next:
            if current.next.data == target:
                # 삭제 관련 
                current.next = current.next.next
                return

            current = current.next
```

```python
# 스택 (Stack) : 노드가 쌓인다!!
# 쌓인다
# 후입선출 (Last In First Out) : LIFO

class Stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top is None

    def push(self, data):
        new_node = Node(data)

        if self.is_empty():
            self.top = new_node
        else:
            current = new_node
            current.next = self.top
            self.top = current

    def display(self):
        if self.is_empty():
            return

        current = self.top

        while current:
            print(f"|  {current.data}  |")
            current = current.next

        print("--------")

    def peek(self):
        # top node 데이터 반환!!
        if self.is_empty():
            return None

        return self.top.data

    def pop(self):
        # top node 데이터 반환 + top node 제거!!
        if self.is_empty():
            return None

        # 1)
        current = self.top
        self.top = self.top.next
        return current.data

        #2)
        #temp = self.top.data
        #self.top = self.top.next
        #return temp
        
# 자료구조 Stack 명세

# Node 활용해서 만들 예정

# 가장 위에 있는 노드는 항상 self.top이 가리킨다

# 함수 push(data)를 통해 노드를 추가한다
# 함수 is_empty()를 통해 비어 있으면 True, 비어 있지 않으면 False 반환
# 함수 display()를 통해 모든 노드 data 출력!!
# 함수 peek()를 통해 top 데이터 확인(반환)
# 함수 pop()을 통해 top 데이터 확인(반환) -> top 노드 제거!!
#    -> top 바로 아래 노드가 top이 된다!!
```

```python
s = Stack()
s.push(10)
s.push(20)
s.push(30)

s.display()

print(s.peek())
```

```python
s2 = Stack()
s2.push(10)
s2.push(20)
s2.push(30)
s2.push(40)
print(s2.peek()) # 40
s2.pop()
print(s2.pop()) # 30
s2.display()
```
