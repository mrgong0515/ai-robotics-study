```python
# 네이버 이미지 사이트에 "로봇팔" 검색 후
# 이미지 500장 다운로드

from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time # 페이지 중간 딜레이주기 위해 시간 관련 라이브러리
import os   # 폴더를 만들자

# keyward = input("검색어를 입력하세요")
keyward = "로봇팔"

if not os.path.exists(keyward): # 키워드 이름으로 된 폴더가 없으면
    os.mkdir(keyward)           # 해당 키워드로 폴더를 만들어라

driver = wb.Chrome()
url_img = "https://search.naver.com/search.naver?where=image"
driver.get(url_img)

time.sleep(0.2) # 0.2초 코드 지연
search_img = driver.find_element(By.ID, "nx_query") # 검색창 태그의 ID 가져오기
search_img.send_keys(keyward)                       # 검색창에 keyward 입력하기 (로봇팔)
search_img.send_keys(Keys.ENTER)
time.sleep(0.2) # 엔터 친 후 이미지 로딩 기다리기
```

```python
# 주의할 점
# 1. 페이지가 이동되면 그 전 까지 가지고 있었던 인스턴스 정보가 초기화 된다 (예 : search_img)

# img 태그 가져오기
imgs = driver.find_elements(By.TAG_NAME, "img")
# <img src = "이미지 주소"></img>
# 이미지를 다운로드 받기 위해 이미지 경로가 필요
# 단, data:로 시작하면 -> 인코딩 필요
# data:가 아닌 친구들만 가져오도록 합시다
import requests # 영화진흥위원회 get 방식으로 요청

count = 0

for img in imgs:
    img_path = img.get_attribute("src")
    if img_path.startswith("data:"):
        continue    # 반복문 이번 회차 진행 ㄴㄴ, 다음 회차로 ㄱㄱ
    
    response = requests.get(img_path)
    # get 방식으로 요청할 때, 응답 포맷은 3가지
    # 1) text, 2)json, 3) byte(content)
    data = response.content
    with open(f"{keyward}/{keyward}{count}.jpg", "wb") as f:    #wb(write byte)
        f.write(data)
        print(f"{keyward}{count}.jpg 저장 완료")
        time.sleep(0.1)
    count += 1
```

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time #
import os   
import requests 

keyward = input("검색어를 입력하세요")
# keyward = "로봇팔"

if not os.path.exists(keyward):
    os.mkdir(keyward)          

driver = wb.Chrome()
url_img = "https://search.naver.com/search.naver?where=image"
driver.get(url_img)

time.sleep(0.2)
search_img = driver.find_element(By.ID, "nx_query") 
search_img.send_keys(keyward)                      
search_img.send_keys(Keys.ENTER)
time.sleep(0.2) 


body = driver.find_element(By.TAG_NAME, "body")
for _ in range(20):
    body.send_keys(Keys.END)
    time.sleep(0.1)

imgs = driver.find_elements(By.TAG_NAME, "img")
count = 0

for img in imgs:
    img_path = img.get_attribute("src")
    if img_path.startswith("data:"):
        continue  
    
    response = requests.get(img_path)
    data = response.content
    with open(f"{keyward}/{keyward}{count}.jpg", "wb") as f: 
        f.write(data)
        print(f"{keyward}{count}.jpg 저장 완료")
        time.sleep(0.05)
    count += 1
driver.close()
```
