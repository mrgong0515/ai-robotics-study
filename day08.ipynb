```python
# Qt UI를 만드는 라이브러리
# C++ QT -> 객체 지향 프로그래밍
from PyQt5.QtWidgets import (QApplication, QWidget, QPushButton, QLabel, QLineEdit)
import sys # 해당 App을 우리 윈도우에 띄워보자!!

# 객체 지향 프로그래밍
# 상속 :검증된 클래스 사용 가능
# 예) 포유류 <- 코끼리
class MyApp(QWidget):
    def __init__(self):
        # 부모 안에 있는 멤버 가져다 쓰자!!
        super().__init__()
        self.initUI()
    
    def initUI(self):
        self.setWindowTitle("First App")
        # 위치, 사이즈 -> self.setGeometry(x, y, width, height)
        self.setGeometry(400, 300, 400, 200)
        # self.move(300, 250)
        # self.resize(400, 200)

        self.btn = QPushButton("눌러라",self)
        self.btn.move(100, 80)
        self.btn.resize(self.btn.sizeHint()) # 버튼 크기 알아서 결정

        # 버튼이 눌러졌을 때 (이벤트가 발생했다!!) 
        # 버튼 이벤트 : 버튼 누름(Press), 버튼 뗌(Release), 눌렀다 뗌(Click), 버튼의 상태가 바뀜
        self.btn.clicked.connect(self.setBGColor)

        # 새로운 버튼 만들어서 배경 색깔 핫핑크
        btn2 = QPushButton("버튼2", self)
        btn2.move(100, 120)
        # btn2는 외부에서 바라볼 일이 없음
        # lambda expression 함수를 정의하지 않고 생성과 동시에 호출 -> 버림
        # btn2.clicked.connect(lambda : self.setStyleSheet("background-color: hotpink"))
        btn2.pressed.connect(lambda : self.setStyleSheet("background-color: red"))
        btn2.released.connect(lambda : self.setStyleSheet("background-color: blue"))
        
        # 텍스트 출력하는 QLabel
        # 배치... 키보드로 배치
        self.label = QLabel("로그인중", self)
        self.label.move(40, 40)

        self.edit_id = QLineEdit("", self)
        self.edit_id.setPlaceholderText("ID:")
        self.edit_pw = QLineEdit("", self)
        self.edit_pw.setPlaceholderText("PW:")

        # pw editor에서 엔터 치면 self.login() 호출
        self.edit_pw.returnPressed.connect(self.login)

        self.edit_id.move(240, 60)
        self.edit_pw.move(240, 100)

        self.btn_login = QPushButton("로그인", self)
        self.btn_login.move(280, 140)

        # 1) 함수 login() 만들어서
        # edit_id와 edit_pw에 있는 문자열 가져오기
        # 미리 지정한 id와 pw면 성공 로직
        # 아니면 실패 로직

        # 2) btn_login과 login() 연결!!
        self.btn_login.clicked.connect(self.login)
        # 추가) 로그인 실패 시 self.label을 "로그인 실패"

        # 로그인 성공 시 edit_id와 edit_pw 내용을 clear

        self.show()
    
    def login(self):
        id = self.edit_id.text()
        pw = self.edit_pw.text()
        if id == "sungwoo" and pw == "54321":
            self.label.setText(f"Welcome {id}!\n 환영합니다!")
            self.label.resize(self.label.sizeHint())
            self.edit_id.setText(None)
            self.edit_pw.setText(None)

        else:
            self.label.setText(f"ID 혹은 PW를 확인하세요.")
            self.label.resize(self.label.sizeHint())

    
    def setBGColor(self):
        # btn이 눌러졌을 때, 실행되는 부분!!
        # self -> 가장 바깥 요소
        self.setStyleSheet("background-color: #00FF00")   # <-rgb 이용한 초록 컬러 설정
        # 버튼의 문구 수정!!
        self.btn.setText("눌렀네?")

# flask server 만들 때!!
# 여러분들이 Python을 사용하는 가장 큰 이유? : 모듈화
# .py -> 자유롭게 import 가능
# Python 인터프리터 언어
#  -> 우리가 import를 진행하면 해당 py파일이 전부 로드
# 개발자가(배포자)가 실행하길 원하지 않는 부분 -> __name__ 활용
# 내가 직접 실행하는 방법(run), import를 통해 실행하는 방법
# 내가 직접 run -> __name__이 __main__으로 바뀜
# __name__ -> day08, __main__ -> day08
# 외부에서 exam.py에서 import day08을 했다
# __name__ -> exam, __main__ -> day08
# 외부에서 import 했을 때, 실행되어야 되지 않는 크리티컬한 코드들은
# __name__으로 처리하자!!

if __name__ == "__main__":
    # 만약, 윈도우 창이 떠 있으면 이전 창 사용!!
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())  # 윈도우 창이 꺼지면 app도 꺼버리자
```

```python
# 기본 코드
from PyQt5.QtWidgets import (QApplication, QWidget, QPushButton, QLabel, QLineEdit,
                             QVBoxLayout, QHBoxLayout, QGridLayout)
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
    
    def initUI(self):

        # 계산기 만들어보자
        vbox = QVBoxLayout()
        grid = QGridLayout()
        self.label = QLabel("0")
        
        vbox.addWidget(self.label)
        vbox.addLayout(grid)

        btn_names = [["7", "8", "9", "*"],
                     ["4", "5", "6", "-"],
                     ["1", "2", "3", "+"],
                     ["C", "0", "=", "/"]]
    
        self.result = ""

        # grid에 버튼 16개 만들어서 넣어보자
        for i in range(4):
            for j in range(4):
                btn = QPushButton("")
                btn.setText(btn_names[i][j])
                grid.addWidget(btn, i, j)
                btn.clicked.connect(self.cal)

        self.setLayout(vbox)
        self.show()

    def cal(self):
        # 어떤 버튼에서 날 호출했는지!!
        btn = self.sender() # 특정 위젯에서 호출 시 해당 위젯 정보를 반환
        key = btn.text()

        if key == "C":
            self.result = ""
            self.label.setText("0")
        elif key == "=":
            answer = str(eval(self.result))
            self.result = answer
            self.label.setText(answer)
        else:
            self.result += key
            # 주의 사항 label.setText(문자열) 숫자 ㄴㄴ
            self.label.setText(self.result)
            
        # 1) 숫자, +, -, *, / 누르면 
        # self.result에 해당 문자열 누적 합!!
        # self.label에 내용 갱신 self.result로

        # 2) C 누르면
        # self.result 비우고 self.label은 0으로

        # 3) = 누르면
        # eval()통해 계산한 결과를 self.label에 출력

        # print(key, end=" ")

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```

```python
# 난이도 버전

from PyQt5.QtWidgets import (QApplication, QWidget, QPushButton, QLabel, QLineEdit,
                             QVBoxLayout, QHBoxLayout, QGridLayout)
import sys
import time

class MyApp(QWidget):
    def __init__(self):
        super().__init__()

        self.mode = 4 # mode x mode 느낌 난도

        self.initUI()
    
    def initUI(self):

        # 1) 기본 UI 틀 잡기
        vbox = QVBoxLayout()
        
        self.label = QLabel(f"1 to {self.mode ** 2}")
        self.btn_start = QPushButton("GAME START")
        grid = QGridLayout()

        vbox.addWidget(self.label)
        vbox.addWidget(self.btn_start)
        vbox.addLayout(grid)

        self.btn_list = []

        for i in range(self.mode):
            for j in range(self.mode):
                btn = QPushButton("")
                btn.setMinimumSize(100, 60)
                btn.setEnabled(False) # 버튼 누르지 못하게!!
                self.btn_list.append(btn)
                grid.addWidget(btn, i, j)

        self.btn_start.clicked.connect(self.start_game)
        self.setLayout(vbox)
        self.show()

    # GAME START 버튼 눌렀을 때
    # 1 ~ 9 까지의 숫자를 각 btn.setText(str(숫자))

    # temp의 요소를 btn에 추가
    # 1) enumerate -> index를 활용
    # 2) zip(temp, self.btn_list)
    # 3) pop()
    def start_game(self):
        temp = list(range(1, self.mode ** 2 + 1))
        import random
        random.shuffle(temp)

        for btn in self.btn_list:
            btn.setEnabled(True) # 모든 버튼 활성화!!
            btn.setText(str(temp.pop()))
            btn.clicked.connect(self.game)

        self.answer = 1
        self.start_time = time.perf_counter()

    def game(self):
        # 1, 2, 3버튼을 눌렀을 때 호출되는 함수
        btn = self.sender()
        key = btn.text()
        print(key, end=" ")

        if int(key) == self.answer:
            btn.setEnabled(False)
            btn.setText("")
            self.answer += 1

            if int(key) == self.mode ** 2:
                end_time = time.perf_counter()
                record = f"{end_time - self.start_time: .2f}초"
                self.label.setText(record)

        # 순차적으로 1, 2, 3, 4, 눌렀을 때
        # 정답을 눌렀을 때, 해당 버튼을 비활성화





if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```

```python
from PyQt5.QtWidgets import (QApplication, QWidget, QPushButton, QLabel, QLineEdit,
                             QVBoxLayout, QHBoxLayout, QGridLayout, QListWidget)
import sys

class MyApp(QWidget):
    def __init__(self):
        super().__init__()
        self.initUI()
    
    def initUI(self):
        self.resize(400, 600)

        vbox = QVBoxLayout()
        self.view = QListWidget()
        self.edit = QLineEdit()
        btn = QPushButton()
        vbox.addWidget(self.view)
        vbox.addWidget(self.edit)
        vbox.addWidget(btn)

        # 숫자 야구 게임
        # 랜덤한 3자리 숫자를 설정 (각 자리는 중복 X)
        # [0 ~ 9] shuffle 된 리스트에서 3개 팜! -> self.answer에 담기

        import random
        temp = list(range(10))
        random.shuffle(temp)
        digit_1 = temp.pop()
        digit_2 = temp.pop()
        digit_3 = temp.pop()
        self.answer = (f"{digit_1}{digit_2}{digit_3}")

        # self.edit에 123 입력 후 btn 클릭하면
        # 해당 숫자가 자리까지 일치 -> strike
        # 자리 일치 X, 숫자는 있다 -> Ball
        # 해당 숫자 X -> Out
        # 사용자에게 123 1S 1B 1OUT 알려준다 (QListWidget에 추가)
        # 3S 나오면 종료!! (축하합니다)

        self.setLayout(vbox)
        self.show()
    def confirm(self):
        guess = self.edit.text()
        strike, ball, out = 0, 0, 0
        # 첫 번째 자리가 Strike, Ball, Out
        if guess[0] == self.answer[0]:
            strike += 1
        elif guess[0] == self.answer[1] or guess[0] == self.answer[2]:
            ball += 1
        else:
            out += 1

if __name__ == "__main__":
    app = QApplication.instance()
    if app is None:
        app = QApplication(sys.argv)
    exam = MyApp()
    sys.exit(app.exec_())
```
