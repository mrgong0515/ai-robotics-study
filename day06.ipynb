```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()

# iris 데이터 셋의 키
print(iris_dataset.keys())

# iris 품종
print("품종 :", iris_dataset["target_names"])

# input 특징
print("특징 이름 :", iris_dataset["feature_names"])

# data의 종류
print("데이터 타입 :", type(iris_dataset["data"]))

# data의 크기
print("데이터 크기 :", iris_dataset["data"].shape)

# 첫 5개 데이터
print("첫 5개 데이터 :\n", iris_dataset["data"][:5])

# Label
print("Label : \n", iris_dataset["target"])

# Lable 데이터 타입
print("Label 데이터 타입 :", type(iris_dataset["target"]))

# Label 개수
print("Label 개수 :", iris_dataset["target"].shape)
```

```python
# 150개의 x(Input)
# 150개의 y(Output)
# 학습

# 모델이 잘 예측하는지 확인!! : 평가
# 150개

# 데이터를 분할
# 학습 : 테스트 = 0.8 : 0.2 ~ 0.7 : 0.3
# 148개는 학습
# 테스트 : 정답이 있는 데이터를 모델에 넣어 예측한 결과와 실제 정답간 비교
# (2개의 데이터 -> input -> 모델 y와 내가 진짜 가지고 있는 정답과 비교)

# 새로운 데이터 ??

# 학습 : 테스트 데이터를 분리
x = iris_dataset.data
y = iris_dataset.target

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=20)
```

```python
# KNN 최근접 이웃 알고리즘
from sklearn.neighbors import KNeighborsClassifier

# 나랑 가까운 이웃 3명을 보고
# 나를 결정 (예: 강아지2, 고양이1 -> 나는 강아지다)
knn = KNeighborsClassifier(n_neighbors=3)  -# -> 이웃의 숫자는 홀수로 해주는게 좋다
# 학습 fit
# 학습 시 주의사항!!
# 절대... Test Dataset을 건드리면 안됩니다 --> Test라는 키워드 금지
knn.fit(x_train, y_train)
```

```python
# 학습된 모델을 가지고 평가해보자!!
# Test 데이터셋을 가지고

# 학습된 knn모델에 x_test를 넣어서 예측 결과를 y_pred에 저장
y_pred = knn.predict(x_test)
# 우리가 만든 모델을 통해 예측한 결과
print(f"우리가 만든 모델을 통해 예측한 결과 : \n{y_pred}")
print(f"정답 결과 : \n{y_test}")

# y_pred와 y_test의 값을 비교하여
# 얼마나 정확한지(accuracy) 확인해보자

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f"정확도 : {accuracy * 100 : .2f}%")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

accuracies = []

for i in range(3, 30, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy = round(accuracy_score(y_test, y_pred), 2)
    accuracies.append(accuracy)

print(accuracies)
```

```python
# 모델의 성능

# 과소적합

# 예를 들어 정확도가 0.7 -> 모델의 예측확률이 낮다
# 규칙, 패턴 단순해서 예측하기 어려울 수 있다
# 학습 성능 낮다, 테스트 성능 낮다
# -> 과소적합(Underfitting)

# 과대적합

# 랜덤 스테이트 : 42 -> 100% 정확도
# 데이터셋은 기가 막히게 맞추는데
# 실 데이터(test data)는 잘 맞추지 못함
# -> 과대적합(Overfitting)
# 학습 성능 높다, 테스트 성능 낮다

# 일반화(Generalization)
# 모델이 학습, 테스트도 잘 통과

# 과소적합 -> feature를 추가!! 더 복잡한 모델 사용
# 과대적합 -> 데이터 정제, Feature 제한, 교차-검증
```

```python
# 2번째 분류 모델 (결정 트리)
# 의사 결정 트리(Decision Tree)
# feature에 대한 질문을 통해 데이터를 분류
# 정보 획득이 최대가 되는 방향으로 feature에 대한 질문을 한다!

# 핵심 개념
# 불순도(Impurity) : 한 노드에 여러 클래스가 얼마나 섞여 있는지 나타내는 정도
# gini 불순도 : 클래스가 섞여 있지 않을 수록 0에 수렴한다
# 엔트로피 : 클래스가 얼마나 안정되어 있는지
# 정보 획득(Information Gain) : 특정 질문을 통해 데이터를 나누었을 때
#                               불순도가 얼마나 감소하는지

# 장점
# 시각화 ! : 눈으로 보인다 !!
# 사람의 언어로 표현하기 용이하다
# 전처리가 쉽다

# 단점
# 과대적합 확률이 굉장히 높다 -> max_depth 제한, 하나의 노드에 들어가는 최소 샘플의 개수 제한 ==> 과대적합 방지할 수 있다.
# train_data에 영향을 많이 받는다
#  -> split을 따라 모델 성능이 크게 좌우된다


# 하나의 상자에 공이 10개 들어있다
# 빨간 공 5개, 파란 공 5개
# 2개의 공을 꺼내면 -> 둘 다 빨강 (22.5%), 둘 다 파랑 (22.5%)    ////     빨강, 파랑 (55%)

# 클래스가 순수 (Pure)
# 빨간 공 10개
# 2개의 공을 꺼내면 -> 둘 다 빨강 (100%)
 
 
 

from sklearn.datasets import load_wine                         # 와인 품종 데이터
from sklearn.tree import DecisionTreeClassifier                # 의사 결정 트리 분류 모델
from sklearn.model_selection import train_test_split           # 훈련, 테스트 분리
from sklearn.metrics import accuracy_score                     # 평가지표 : 정확도

wine_dataset = load_wine()

# print("key : ", wine_dataset.keys())
# print("data : ", wine_dataset.data[:2])
# print("target : ", wine_dataset.target)
# print("frame : ", wine_dataset.target_names)
# print("feature_names : ", wine_dataset.feature_names)


xx = wine_dataset.data
yy = wine_dataset.target
xx_train, xx_test, yy_train, yy_test = train_test_split(xx, yy, test_size= 0.25, random_state= 20)
tree = DecisionTreeClassifier(max_depth=3, random_state=20)

tree.fit(xx_train, yy_train)

yy_pred = tree.predict(xx_test)
accuracy = accuracy_score(yy_test, yy_pred)
print(f"정확도 : {accuracy * 100: .2f}%")
```

```python
# 의사 결정 트리 시각화
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(32,24))
plot_tree(tree, feature_names=wine_dataset.feature_names, filled=True, class_names=wine_dataset.target_names)
plt.show()
```

```python
import pandas as pd
# 어떤 feature가 영향을 가장 많이 미치는가?
tree_importance = tree.feature_importances_
data = {"feature":wine_dataset.feature_names, "importance":tree_importance}
df_wine = pd.DataFrame(data)
df_wine.sort_values(by=["importance"],ascending=[False])
```

```python
# 유방암 데이터를 의사 결정 트리로 분류해보자

# 1) 의사결정나무 클래스 import
# 2) dataset
from sklearn.datasets import load_breast_cancer
# 3) train_test_split import
# 4) accuracy import

# load_breast_cancer를 통해 cancer 인스턴스 생성
cancer = load_breast_cancer
# cancer.keys()를 통해 어떤 데이터 타입, 개수, 클래스 이름 등 확인
cancer.keys() # 'data', 'target', 'frane', 'target_names', 'DESCR', 'feature_names', 'filename', 'data_module'

# 6) x, y 설정
x = cancer.data
y = cancer.target
# 7) train과 test 분할(분리)
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=20)
# 8) 의사결정나무 클래스를 통해 d_tree 인스턴스 생성, 단 max_depth=3, random_state=20
d_tree = DecisionTreeClassifier(max_depth=3, random_state=20)
# 9) fit(학습)
d_tree.fit(x_train, y_train)
# 10) 학습된 모델에 x_test 넣어 y_pred 예측!!
y_pred = d_tree.predict(x_test)
# 11) y_pred, y_test 비교해서 정확도(accuracy) 측정
accuracy_score(y_test, y_pred)
print(f"정확도 : {accuracy * 100 : .2f}%")
# 12) plot_tree 통해 시각화!!
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plt.figure(figsize=(24, 16))
plot_tree(d_tree, filled=True, feature_names=cancer.feature_names, class_names=cancer.target_names)
plt.show()
# 13) 어떤 요인이 영향을 미쳤는지?(importance) 확인!!
import pandas as pd
tree_importance = d_tree.feature_importances_
data = {"feature":cancer.feature_names, "importance":tree_importance}
df_cancer = pd.DataFrame(data)
df_cancer.sort_values(by=["importance"], ascending=[False])
```
